```python
import pandas as pd
import numpy as np

import os
```

```python
lines = pd.read_csv(
    "movie-conversation/movie_lines.tsv", 
    encoding='utf-8-sig', 
    sep='\t', 
    on_bad_lines="skip", 
    header = None,
    names = ['lineID', 'charID', 'movieID', 'charName', 'text'],
    index_col=['lineID']
)
```

```python
lines
```

```python
data = {}

for d in lines.itertuples():
    if type(d[4]) != type(''):
        continue

    words = list(filter(lambda x: x != '', d[4].split(' ')))
    name = d[1].lower()
    try:
        data[name] += words
    except:
        data.__setitem__(name, words)
```

```python
df = pd.DataFrame({
    'character': list(data.keys()),
    'words': data.values() 
})
```

```python
df.to_csv('movie-conversation.csv', index=False)
```
